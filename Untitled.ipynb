```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('zomato.csv')
```

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis=1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
#cleaning rate column

df['rate'].unique()
```

```python
#remove string values ,we will remove rows, /5 we need to remove

def handlerate(value):
    if (value == 'NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.isnull().sum()
```

```python
#filling rate values with mean

df['rate'].fillna(df['rate'].mean(), inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
#renaming

df.rename(columns = {'approx_cost(for two people)': 'Cost2plates' , 'listed_in(type)': 'Type'}, inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df.drop(['listed_in(city)'], axis=1)
```

```python
df.head()
```

```python
df['Cost2plates'].unique()
```

```python
#removing comma from Cost2places

def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending= False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if (value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
location = df['location'].value_counts( ascending = False)
location_lessthan300 = location[location<300]

def handle_location(value):
    if (value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
#cleaning cuisines

cuisines = df['cuisines'].value_counts(ascending = False)

cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if (value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
#Data is clean, now go for visualisation
```

```python
#count plot for various location

plt.figure(figsize=(16, 10))
location_counts = df['location'].value_counts()
ax = sns.barplot(x=location_counts.index, y=location_counts.values)
plt.xticks(rotation=90)
```

```python
# Assuming 'online_order' column contains string values 'Yes' and 'No'
# Convert 'Yes' to 1 and 'No' to 0
df['online_order'] = df['online_order'].map({'Yes': 1, 'No': 0})

plt.figure(figsize=(6, 6))
sns.countplot(x='online_order', data=df, palette='inferno')
```

```python
#visualising book table facility 

df['book_table'] = df['book_table'].map({'Yes': 1, 'No': 0})

plt.figure(figsize=(6, 6))
sns.countplot(x='book_table', data=df, palette='rainbow')
```

```python
# visualizing online order vs rate

plt.figure(figsize = (6,6))
sns.boxplot(x='online_order', y='rate', data=df)
```

```python
# visualizing book table vs rate

plt.figure(figsize = (6,6))
sns.boxplot(x='book_table', y='rate', data=df)
```
